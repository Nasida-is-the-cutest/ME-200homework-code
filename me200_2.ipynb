```python
#HW 2.3
import numpy as np

A = 0.800 # cm^2
m = 25.000 # kg
ΔU = 0.100 # kJ
ΔPE = 0.200 # kJ
F = 1334.000 # N


D = 0.1 # m
ΔU = ΔU*1000 # J
ΔPE = ΔPE*1000 # J
s = D**2/4 * np.pi # m^2
g = 9.810 # m/s^2
p_atm=10**5 # Pa    
A=A*1e-4 # m^2

p=(F+m*g)/s + p_atm # Pa
Δh=ΔPE/(m*g) # m
w_shaft=-F*Δh # J
print("w_shaft", -w_shaft)

w_atmosphere=-p_atm*(s-A)*Δh # J  #只有s-A的面积部分对大气做功，极易忘记
print("w_atmosphere", -w_atmosphere)

q=ΔU +ΔPE- w_shaft - w_atmosphere # J
print("q", q)
```
